HEAT MAP

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg')
```

```python
cars.head()
```

```python
cars.groupby('origin').cylinders.value_counts() #summarize date using groupby
```

```python
cars.groupby('origin').cylinders.value_counts().unstack().fillna(0) #unstack data for better view , fill null values will 0
```

```python
origin_cyl = cars.groupby('origin').cylinders.value_counts().unstack().fillna(0)
```

```python
sns.heatmap(origin_cyl)
```

```python
sns.heatmap(origin_cyl) #Transpose (switch rows and columns) for any numpy array or pandas df
```

```python
sns.heatmap(origin_cyl.T)
```

```python
sns.heatmap(origin_cyl, cmap = 'Blues') #colormap(cmap) helps adjust color pallets in seaborn
```

```python
sns.heatmap(origin_cyl, cmap = 'Blues', annot = True, fmt='.0f') #annot shows values each box, fmt changes nos to integers
```

```python
sns.heatmap(origin_cyl, cmap = 'Blues', annot = True
            , fmt='.0f', annot_kws = {'fontsize':16, 'fontweight':'bold', 'fontfamily':'serif'})
```

REGPLOTS aka Regression Plots

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')
```

```python
diamonds.shape
```

```python
diamonds. head()
```

```python
diamonds = diamonds.sample(n = 200, random_state = 44) #randomly selects 200 datapoints from diamonds datasetsns
diamonds.shape
```

```python
sns.set_style('dark')
```

```python
sns.regplot(diamonds.carat, diamonds.price) #finding correlation between carat and price variables
```

```python
sns.regplot(diamonds.carat, diamonds.price, fit_reg = False) #fit_reg takes out the line
```

```python
sns.regplot(diamonds.carat, diamonds.price, scatter = False) #scatter = False takes out scatter plot
```

```python
sns.regplot(diamonds.carat, diamonds.price, ci = None) #ci = None turns off confidence interval (shaded line portion)
```

```python
sns.regplot(diamonds.carat, diamonds.price, order = 2) #order =2 implies a polynomical distr giving the curve line effect
```

```python
sns.regplot(x='carat', y='price', data=diamonds,  marker='d') #marker changes shape of regression dots
```

```python
sns.regplot(x='carat', y='price', data=diamonds,
            scatter_kws={'s': 100, 'alpha': 0.5, 'color': 'lightgray'}  #s = size of dots, alpha = transparency
            ); #scatter keywords introduces new arguments to modify scatter plot
```

```python
sns.regplot(x='carat', y='price', data=diamonds,
            ci=None,
            line_kws={'lw': 4, 'color': 'black', 'linestyle': '-.'}
            ); #line keywords introduces new arguments to modify line
```
