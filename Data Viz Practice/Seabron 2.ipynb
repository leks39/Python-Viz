```python
##DISPLOT - distribution plot 
```

```python
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna() #dropna() 
```

```python
cars.head()
```

```python
cars.shape
```

```python
sns.displot(cars.weight)
```

```python
sns.displot(cars.weight, kde = True) #adds kde line
```

```python
sns.displot(cars.weight, kind = 'ecdf') #kind specifies type of distr which could be hist(default), kde, ecdf
```

```python
sns.displot(cars.weight, kind = 'kde', rug = True) #rug below to see where distributions are falling
```

```python
sns.displot(x='weight', data=cars, hue='origin'); #hue splits data based on a particlar categorical variable
```

```python
sns.displot(x='weight', data=cars, hue='origin', col = 'origin'); #col creates a new column for each categorical variable
```

```python
sns.displot(x='weight', data=cars, hue='origin', col = 'origin', kind = 'kde'); #kind be introduced with col
```

```python
sns.displot(x='weight', data=cars[cars.cylinders.isin([4, 6, 8])], col='origin',  #isin selects only 4,6,8 
            row='cylinders', 
            hue='cylinders'); 
```

```python
sns.displot(x='weight', data=cars, hue='origin', col='origin', palette='crest'); #palette changes color format
```

```python
sns.displot(x='weight', data=cars, hue='origin', col='origin', height = 3) #default height of bars = 5, can be changed
```

```python
sns.displot(x = 'weight', data = cars, 
            hue = 'origin', col = 'origin', height = 2, aspect = 1.5) #aspect measures width/height, default =1 
```
